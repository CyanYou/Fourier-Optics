# Fourier Optics - Investigating the performance of a simple imaging system

## Introduction

There are a number of ways of characterising the performance and resolution of an optical imaging system.  The one you are probably most familiar with considers the point spread function (PSF) of the system and it is this approach that leads us to measures, such as Rayleigh’s criterion, that tell us how close two points can be such that they are still resolvable.  You will also have already carried out investigations that demonstrate the degradation of the PSF in the presence of aberrations particularly from a ray-tracing viewpoint.  

An alternative approach was first introduced by Ernst Abbe and is based around the concept of how simple grating patterns are imaged by an optical system - the higher the resolution of an optical system, the finer the grating pattern it can image.  We can think of any object being made up of a combination of these simple grating patterns, or spatial frequencies, which can be mathematically determined by Fourier analysis.  The effect of the optical system on imaging each of these individual Fourier components can be calculated and then the resulting outputs can be summed to determine what the image looks like.  It is this complementary approach that is at the heart of Fourier optics and which is investigated in this experiment.

The alternative approach to thinking about optical resolution is to consider how an imaging system performs when viewing an object consisting of a single spatial frequency.  What this means is simply a grating or grid pattern that has a sinusoidal amplitude or intensity profile.  The resulting measure of resolution is usually given in terms of the maximum line pairs per millimetre (or spatial frequency) that can be resolved.  Any intensity or amplitude object or image can be thought of as a summation of different spatial frequency components in the same way that a time varying signal can be considered as the summation of a number of temporal frequency components (sine and cosine waves).  If we know how the different spatial frequency components propagate through an optical system then we can work out what an image of an object will look like.  

### Spatial coherence

One concept that is key to the Fourier analysis of optical systems is that of spatial coherence.  A spatially coherent system is often characterised by illumination from a point source or a laser and you will be able to observe diffraction and interference effects (such as speckle).  A spatially incoherent system is often illuminated by an extended or diffuse source such as daylight or a light bulb and you will not in general be able to observe significant diffractive or interference effects.  Essentially, spatial coherence is a measure of the correlation in phase between light from different points on the object, i.e. whether they can interfere to produce an interference pattern or not.  \[The similar concept of temporal coherence tells us whether light arriving after travelling different times is correlated and can interfere.\]  You may already have seen some of the effects of spatial coherence in the microscope illumination demonstration experiment by observing changes in the image (ringing at edges) as the condenser aperture is closed down.

The degree of spatial coherence in an imaging system determines how we deal with it in terms of a Fourier optics approach.  A spatially coherent system is linear in amplitude and we can therefore use Fourier analysis on the amplitude image and its spatial frequency components whereas a spatially incoherent system is linear in intensity and we must consider the intensity image and its spatial frequency components.  Unfortunately there is also a third option when the imaging is partially coherent – this occurs when the source is of an intermediate size and such systems are linear in neither amplitude nor intensity and must be treated very carefully!

# Spatially incoherent imaging

Spatially incoherent imaging is the case when the source iris is bigger than the aperture iris (typically a factor of $S=2$ is considered spatially incoherent.  You can consider the larger source iris to be a combination of lots of smaller irises translated latterally as you will just have tried out.  However the final image is a summation over the **intensity** images produced by each sub-iris as they will be all mutually incoherent. What this results with is a transfer function (or OTF) that is the **autocorrelation** of the aperture iris function, as is explained further in the appendix.

## Appendix: The frequency response of optical systems – the optical transfer function.

You of course are already aware that an optical imaging system has a finite resolution given by the size of the point spread function of that system, and it is upon this basis that resolution criteria such as Rayleigh’s criterion are based.  However another way of looking at an optical system is to think in terms of its spatial frequency response.  In the same way that an electrical or optical (fibre) system has a temporal frequency response whereby we can determine how a particular frequency is transmitted by the system, we can analyse the performance of an imaging system in terms of the way that spatial frequency components ate transmitted by the system.  Also, in the same way that the temporal frequency response is equivalent to the Fourier transform of the impulse response function, so the spatial frequency response of an optical imaging system is the Fourier transform of the point spread function of that system.  For a thorough treatment of these principles see the classic text “Introduction to Fourier Optics” by Joseph Goodman.

### The frequency response of a spatially coherent system
 
| ![Image](aotf.png) |
|:-:|
| **Figure 4: Geometry for calculating the coherent OTF of a lens** |

It is perhaps easiest to understand the frequency response of a coherent optical system first.  Here we must analyse the performance of the system in terms of amplitudes.  Consider the imaging system as shown in Figure 4 consisting of a simple lens of focal length $f$ with a pupil of diameter $D$ defined in its back focal plane.  A single spatial frequency in the front focal plane is actually a tilted plane wave travelling at an angle $\theta$ to the optic axis.  Its lateral spatial frequency is $\nu=\sin(\theta)/\lambda$ and this wave travels through the lens and is brought to a focus in the back focal plane at a height of $h=f\times\sin(\theta)=f\lambda\nu$ (particularly when the lens satisfies the Abbe sine condition).    Thus spatial frequency directly corresponds to ray height in the back focal plane and the whole plane wave travelling at an angle to the optic axis is focused to that point in the back focal plane of the lens.  The frequency response is therefore determined by whether that ray passes through the pupil or not i.e. if $\nu<D/(2f\lambda)$.  The coherent amplitude optical transfer function (OTF) of a perfect lens is thus simply a circle of radius $D/(2f\lambda)$.

Relating this result back to the amplitude point spread function is now trivial as we already know that the amplitude point spread function of a lens is the Fourier transform of its pupil function.

### The frequency response of a spatially incoherent optical system

This is a little more tricky to visualize than the amplitude system above as we know nothing about amplitudes in a purely incoherent system – we can only measure intensities.  Therefore we cannot think easily in terms of plane waves travelling in different directions.  However we do know that the point spread function for a spatially incoherent system is exactly the same as for the coherent system so long as we consider only its intensity.  The frequency response of the spatially incoherent system is then the Fourier transform of the intensity point spread function.  Using the convolution theorem from Fourier theory and can therefore express the spatially incoherent transfer function as the **autocorrelation** of the pupil function.

| ![Image](iotf.png) |
|:-:|
| **Figure 5: Geometry for calculating the incoherent OTF of an optical system** |

The autocorrelation function is a measure of the overlap of a function with a copy of itself as the copy it shifted.  Figure 6 shows a diagram that will help you to calculate this function for a circular aperture with a maximum coherent spatial frequency cutoff of $\nu_0$.  You should calculate the incoherent optical transfer function using this diagram and show that when properly normalised it is equal to:

$$H(\nu)=\left\{\begin{aligned}\frac{2}{\pi} \left[\cos^{-1}⁡\left(\frac{\nu}{2\nu_0}\right)-\frac{\nu}{2\nu_0}\sqrt{1-\left(\frac{\nu}{2\nu_0}\right)^2}\right]&\;\;\;\;\;\; \nu<2\nu_0 \\
    0 & \;\;\;\;\;\;\mathrm{otherwise}
    \end{aligned}\right.$$

The maximum spatial frequency that is transmitted by the incoherent optical system is $2\nu_0$, which is twice the maximum amplitude frequency transmitted by the coherent system.  \[This may sound alarming at first but consider that when you detect intensity in the coherent system you detect the square of amplitude which will have components at twice the frequency of the original amplitude\]

The incoherent transfer function is plotted below:

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
def otf(kx, ky, kmax):
    kr = np.sqrt(kx ** 2 + ky ** 2) / kmax
    ans = np.zeros_like(kr)
    ans[kr < 1] = 2 / np.pi * (np.arccos(kr[kr < 1]) - kr[kr < 1] * np.sqrt(1 -  kr[kr < 1] ** 2))
    return ans
    
plt.figure()
plt.plot(np.linspace(0, 2, 100), otf(np.linspace(0, 2, 100), 0, 2))
plt.title('incoherent transfer function')
plt.xlabel('normalised spatial frequency')
plt.show()
```

### The frequency response of a partially coherent system

This is harder than in the previous two cases.  The coherent system is linear in amplitude so we can define an amplitude frequency response; the incoherent system is linear in intensity so we can define an intensity frequency response.  Unfortunately a partially coherent system is a mixture of the incoherent and the coherent.  It is therefore linear in neither amplitude nor intensity and we cannot properly define a frequency response for the system.  For any given object we can determine how the spatial frequencies in that particular object are imaged.  We cannot however generalise that those same spatial frequencies will be imaged the same way in all objects!
